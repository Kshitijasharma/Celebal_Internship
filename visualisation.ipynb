```python
# Import necessary libraries
from sklearn.datasets import load_iris
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the Iris dataset from sklearn
iris_sklearn = load_iris()
iris_df = pd.DataFrame(data=iris_sklearn.data, columns=iris_sklearn.feature_names)
iris_df['species'] = pd.Categorical.from_codes(iris_sklearn.target, iris_sklearn.target_names)

# Display the first few rows of the dataset
print(iris_df.head())
```

Pair Plot: Shows the pairwise relationships between features colored by species.

```python
# Pair Plot
sns.pairplot(iris_df, hue='species')
plt.suptitle('Pair Plot of Iris Dataset', y=1.02)
plt.show()
```

Box Plot: Displays the distribution of each feature across the entire dataset.

```python
# Box Plot
plt.figure(figsize=(12, 6))
sns.boxplot(data=iris_df, orient="h")
plt.title('Box Plot of Iris Dataset')
plt.show()
```

Violin Plot: Illustrates the distribution and density of each feature for each species.

```python
# Violin Plot
plt.figure(figsize=(12, 6))
sns.violinplot(x="species", y="sepal length (cm)", data=iris_df)
sns.violinplot(x="species", y="sepal width (cm)", data=iris_df)
sns.violinplot(x="species", y="petal length (cm)", data=iris_df)
sns.violinplot(x="species", y="petal width (cm)", data=iris_df)
plt.title('Violin Plot of Iris Dataset')
plt.show()
```

Heatmap: Visualizes the correlation between the features in the dataset.

```python
# Heatmap
plt.figure(figsize=(10, 8))
correlation_matrix = iris_df.iloc[:, :-1].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Feature Correlations in Iris Dataset')
plt.show()
```

Histogram: To show the distribution of each feature.

```python
# Histogram
iris_df.hist(figsize=(12, 8), bins=20, edgecolor='black')
plt.suptitle('Histogram of Iris Dataset Features')
plt.show()
```

Joint Plot: To visualize the relationship between two features with a scatter plot and histograms.

```python
# Joint Plot
sns.jointplot(x='sepal length (cm)', y='sepal width (cm)', data=iris_df, hue='species', kind='scatter')
plt.suptitle('Joint Plot of Sepal Length and Width', y=1.02)
plt.show()

sns.jointplot(x='petal length (cm)', y='petal width (cm)', data=iris_df, hue='species', kind='scatter')
plt.suptitle('Joint Plot of Petal Length and Width', y=1.02)
plt.show()
```
